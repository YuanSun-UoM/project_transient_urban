## CreateVariables

```python
dyn_alb_improad_TBD = f.createVariable('dyn_alb_improad_TBD',np.float64,('time', 'lat', 'lon'),zlib=True,complevel=3)
dyn_alb_improad_HD = f.createVariable('dyn_alb_improad_HD',np.float64,('time', 'lat', 'lon'),zlib=True,complevel=3)
dyn_alb_improad_MD = f.createVariable('dyn_alb_improad_MD',np.float64,('time', 'lat', 'lon'),zlib=True,complevel=3)
```

## SetVariableAttributes

```python
# set variable attributes
dyn_alb_improad_TBD.setncatts({'long_name':'time-varying improad albedo for Tall Building District (TBD) class'})
dyn_alb_improad_HD.setncatts({'long_name':'time-varying improad albedo for High Density (HD) class'})
dyn_alb_improad_MD.setncatts({'long_name':'time-varying improad albedo for Medium Density (MD) class'})
```

## WriteInData

```python
improad_albedo_value = f2.variables['ALB_IMPROAD_DIR'][:,:,:]
```

```python
for i in range (258):
    f.variables['dyn_alb_improad_TBD'][i,:,:] = improad_albedo_value[0,0,:,:]
```

```python
for i in range (258):
    f.variables['dyn_alb_improad_HD'][i,:,:] = improad_albedo_value[0,1,:,:]
```

```python
for i in range (258):
    f.variables['dyn_alb_improad_MD'][i,:,:] = improad_albedo_value[0,2,:,:]
```

## IncreaseYearlyAlbedo
- i = 0 -> 1849
- i = 166 -> 2015
- i = 201 -> 2050
- i = 251 -> 2100

```python
improad_albedo_constantTBD = f.variables['dyn_alb_improad_TBD'][:,:,:]
improad_albedo_constantHD = f.variables['dyn_alb_improad_HD'][:,:,:]
improad_albedo_constantMD = f.variables['dyn_alb_improad_MD'][:,:,:]

for i in range (166,258):
    alb_improad_TBD = f.variables['dyn_alb_improad_TBD'][i,:,:]
    alb_improad_TBD = np.where(alb_improad_TBD > 0 , alb_improad_TBD + (i - 165)*0.01, alb_improad_TBD)
    f.variables['dyn_alb_improad_TBD'][i,:,:] = np.minimum(alb_improad_TBD, 0.9)

for i in range (166,258):
    alb_improad_HD = f.variables['dyn_alb_improad_HD'][i,:,:]
    alb_improad_HD = np.where(alb_improad_HD > 0 , alb_improad_HD + (i - 165)*0.01, alb_improad_HD)
    f.variables['dyn_alb_improad_HD'][i,:,:] = np.minimum(alb_improad_HD, 0.9)

for i in range (166,258):
    alb_improad_MD = f.variables['dyn_alb_improad_MD'][i,:,:]
    alb_improad_MD = np.where(alb_improad_MD > 0 , alb_improad_MD + (i - 165)*0.01, alb_improad_MD)
    f.variables['dyn_alb_improad_MD'][i,:,:] = np.minimum(alb_improad_MD, 0.9)
```
