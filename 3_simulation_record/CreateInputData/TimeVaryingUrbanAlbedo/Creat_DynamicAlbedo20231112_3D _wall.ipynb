## CreateVariables

```python
dyn_alb_wall_TBD = f.createVariable('dyn_alb_wall_TBD',np.float64,('time', 'lat', 'lon'),zlib=True,complevel=3)
dyn_alb_wall_HD = f.createVariable('dyn_alb_wall_HD',np.float64,('time', 'lat', 'lon'),zlib=True,complevel=3)
dyn_alb_wall_MD = f.createVariable('dyn_alb_wall_MD',np.float64,('time', 'lat', 'lon'),zlib=True,complevel=3)
```

## SetVariableAttributes

```python
# set variable attributes
dyn_alb_wall_TBD.setncatts({'long_name':'time-varying wall albedo for Tall Building District (TBD) class'})
dyn_alb_wall_HD.setncatts({'long_name':'time-varying wall albedo for High Density (HD) class'})
dyn_alb_wall_MD.setncatts({'long_name':'time-varying wall albedo for Medium Density (MD) class'})
```

## WriteInData

```python
wall_albedo_value = f2.variables['ALB_WALL_DIR'][:,:,:]
```

```python
for i in range (258):
    f.variables['dyn_alb_wall_TBD'][i,:,:] = wall_albedo_value[0,0,:,:]
```

```python
for i in range (258):
    f.variables['dyn_alb_wall_HD'][i,:,:] = wall_albedo_value[0,1,:,:]
```

```python
for i in range (258):
    f.variables['dyn_alb_wall_MD'][i,:,:] = wall_albedo_value[0,2,:,:]
```

## IncreaseYearlyAlbedo
- i = 0 -> 1849
- i = 166 -> 2015
- i = 201 -> 2050
- i = 251 -> 2100

```python
wall_albedo_constantTBD = f.variables['dyn_alb_wall_TBD'][:,:,:]
wall_albedo_constantHD = f.variables['dyn_alb_wall_HD'][:,:,:]
wall_albedo_constantMD = f.variables['dyn_alb_wall_MD'][:,:,:]

for i in range (166,258):
    alb_wall_TBD = f.variables['dyn_alb_wall_TBD'][i,:,:]
    alb_wall_TBD = np.where(alb_wall_TBD > 0 , alb_wall_TBD + (i - 165)*0.01, alb_wall_TBD)
    f.variables['dyn_alb_wall_TBD'][i,:,:] = np.minimum(alb_wall_TBD, 0.9)
       # f.variables['dyn_alb_wall_TBD'][i,:,:] = wall_albedo_constantTBD[i,:,:] + (i -165)*0.01
       # f.variables['dyn_alb_wall_TBD'][i,:,:] = np.minimum(f.variables['dyn_alb_wall_TBD'][i,:,:], 0.9)

for i in range (166,258):
    alb_wall_HD = f.variables['dyn_alb_wall_HD'][i,:,:]
    alb_wall_HD = np.where(alb_wall_HD > 0 , alb_wall_HD + (i - 165)*0.01, alb_wall_HD)
    f.variables['dyn_alb_wall_HD'][i,:,:] = np.minimum(alb_wall_HD, 0.9)

for i in range (166,258):
    alb_wall_MD = f.variables['dyn_alb_wall_MD'][i,:,:]
    alb_wall_MD = np.where(alb_wall_MD > 0 , alb_wall_MD + (i - 165)*0.01, alb_wall_MD)
    f.variables['dyn_alb_wall_MD'][i,:,:] = np.minimum(alb_wall_MD, 0.9)    

    
```
